```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"
    
    def turn_on(self):
        self.status = "ON"
        print(f"{self.light_name} is {self.status}")
    
    def turn_off(self):
        self.status = "OFF"
        print(f"{self.light_name} is {self.status}")

# Main program
if __name__ == "__main__":
    # Getting input
    light_name = input("Light Name: ")
    action = input("Action: ").upper()
    
    # Creating light object
    light = SmartLight(light_name)
    
    # Performing action
    if action == "ON":
        light.turn_on()
    elif action == "OFF":
        light.turn_off()
    else:
        print("Invalid action. Please enter ON or OFF.")
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
    
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    def display_id_card(self):

        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

# Main program
if __name__ == "__main__":
    # Getting input from user
    name = input("Employee Name: ")
    emp_id = input("Employee ID: ")
    department = input("Department: ")
    
    # Creating employee object
    employee = Employee(name, emp_id, department)
    
    # Displaying ID card
    employee.display_id_card()
```

```python
class Contact:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number
    
    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

# Main program
if __name__ == "__main__":
    # Get input from user
    name = input("Contact Name: ")
    phone = input("Phone Number: ")
    
    # Create contact object
    contact = Contact(name, phone)
    
    # Display contact information
    contact.display_contact()
```

```python
class PriceTag:
    def __init__(self):
        self.name = input("Product Name: ")
        self.price = input("Price: ")
    
    def show(self):
        print(f"\nProduct: {self.name}")
        print(f"Price: ₹{self.price}")

# Creating and display price tag
tag = PriceTag()
tag.show()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    
    def display_movie_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

# Main program
if __name__ == "__main__":
    # Get input from user
    name = input("Movie Name: ")
    rating = input("Rating: ")
    
    # Create movie object
    movie = Movie(name, rating)
    
    # Display movie details
    movie.display_movie_details()
```

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    
    def print_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

# Main program
if __name__ == "__main__":
    # Get input from user
    customer_name = input("Customer Name: ")
    address = input("Address: ")
    
    # Create delivery object
    delivery = Delivery(customer_name, address)
    
    # Display delivery details
    delivery.print_delivery_details()
```
